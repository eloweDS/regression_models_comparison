```python
from sklearn.preprocessing import LabelEncoder

for col in df.columns:
    if df[col].dtype == "object":
        print(str(col))
        label = LabelEncoder()
        label = label.fit(df[col])
        df[col] = label.transform(df[col].astype(str)).astype(int)
df = df.astype({col: 'int' for col in df.select_dtypes(include='object').columns})
```

```python
from sklearn.metrics import mean_squared_error, r2_score
import numpy as np
import pandas as pd
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import StandardScaler, PolynomialFeatures
from sklearn.linear_model import Ridge, Lasso, LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.decomposition import PCA
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
X = df.drop(columns=['target_v'])
y = df['target_v']
results = {
    'Regression Tree': {'MSE_train': [], 'MSE_test': [], 'RMSE_train': [], 'RMSE_test': [], 'RSE_train': [], 'RSE_test': [], 'RSS_train': [], 'RSS_test': [], 'R^2_train': [], 'R^2_test': []},
    'Ridge Regression': {'MSE_train': [], 'MSE_test': [], 'RMSE_train': [], 'RMSE_test': [], 'RSE_train': [], 'RSE_test': [], 'RSS_train': [], 'RSS_test': [], 'R^2_train': [], 'R^2_test': []},
    'Lasso Regression': {'MSE_train': [], 'MSE_test': [], 'RMSE_train': [], 'RMSE_test': [], 'RSE_train': [], 'RSE_test': [], 'RSS_train': [], 'RSS_test': [], 'R^2_train': [], 'R^2_test': []},
    'Principal Component Regression (PCR)': {'MSE_train': [], 'MSE_test': [], 'RMSE_train': [], 'RMSE_test': [], 'RSE_train': [], 'RSE_test': [], 'RSS_train': [], 'RSS_test': [], 'R^2_train': [], 'R^2_test': []},
    'Regression Splines': {'MSE_train': [], 'MSE_test': [], 'RMSE_train': [], 'RMSE_test': [], 'RSE_train': [], 'RSE_test': [], 'RSS_train': [], 'RSS_test': [], 'R^2_train': [], 'R^2_test': []},
}

num_features = X.shape[1]

for _ in range(20):
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=None)

    models = {
        'Regression Tree': DecisionTreeRegressor(max_depth=3),
        'Ridge Regression': Ridge(),
        'Lasso Regression': Lasso(),
        'Principal Component Regression (PCR)': make_pipeline(StandardScaler(), PCA(n_components=min(5, num_features)), LinearRegression()),
        'Regression Splines': make_pipeline(PolynomialFeatures(degree=2), LinearRegression()),
    }
    for model_name, model in models.items():
        model.fit(X_train, y_train)
        y_train_pred = model.predict(X_train)
        y_test_pred = model.predict(X_test)

        mse_train = mean_squared_error(y_train, y_train_pred)
        rmse_train = np.sqrt(mse_train)
        rss_train = np.sum((y_train - y_train_pred) ** 2)
        rse_train = np.sqrt(rss_train / (len(y_train) - X_train.shape[1] - 1))
        r2_train = r2_score(y_train, y_train_pred)

        mse_test = mean_squared_error(y_test, y_test_pred)
        rmse_test = np.sqrt(mse_test)
        rss_test = np.sum((y_test - y_test_pred) ** 2)
        rse_test = np.sqrt(rss_test / (len(y_test) - X_test.shape[1] - 1))
        r2_test = r2_score(y_test, y_test_pred)

        results[model_name]['MSE_train'].append(mse_train)
        results[model_name]['RMSE_train'].append(rmse_train)
        results[model_name]['RSS_train'].append(rss_train)
        results[model_name]['RSE_train'].append(rse_train)
        results[model_name]['R^2_train'].append(r2_train)

        results[model_name]['MSE_test'].append(mse_test)
        results[model_name]['RMSE_test'].append(rmse_test)
        results[model_name]['RSS_test'].append(rss_test)
        results[model_name]['RSE_test'].append(rse_test)
        results[model_name]['R^2_test'].append(r2_test)

average_results = {model: {metric: np.mean(scores) for metric, scores in metrics.items()} for model, metrics in results.items()}
results_df = pd.DataFrame(average_results).T
results_df['MSE_gap'] = results_df['MSE_train'] - results_df['MSE_test']
results_df['RMSE_gap'] = results_df['RMSE_train'] - results_df['RMSE_test']
results_df['RSS_gap'] = results_df['RSS_train'] - results_df['RSS_test']
results_df['RSE_gap'] = results_df['RSE_train'] - results_df['RSE_test']
results_df['R^2_gap'] = results_df['R^2_train'] - results_df['R^2_test']
results_df
```

```python
import matplotlib.pyplot as plt
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.model_selection import train_test_split, GridSearchCV, learning_curve
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import StandardScaler, PolynomialFeatures, SplineTransformer
from sklearn.decomposition import PCA
from sklearn.linear_model import Ridge, Lasso, LinearRegression
from sklearn.tree import DecisionTreeRegressor
import numpy as np
import pandas as pd
import json

# Define hyperparameter grids
param_grids = {
    'Ridge Regression': {'ridge__alpha': [0.01, 0.1, 1, 10]},
    'Lasso Regression': {'lasso__alpha': [0.01, 0.1, 1, 10]},
    'Principal Component Regression (PCR)': {'pca__n_components': [3, 5, 6, 7, 8]},
    'Regression Splines': {
        'splinetransformer__degree': [2, 3, 4],
        'splinetransformer__n_knots': [5, 10, 20]
    },
    'Regression Tree': {
        'decisiontreeregressor__max_depth': [3, 4, 5,6,7],
        'decisiontreeregressor__min_samples_split': [5, 10, 15,20],
        'decisiontreeregressor__min_samples_leaf': [5, 10, 15,20],
        'decisiontreeregressor__max_features': ['sqrt', 'log2', None]
    }
}

# Initialize results dictionary
results = {model: {'Train MSE': [], 'Test MSE': [], 'Train RMSE': [], 'Test RMSE': [],
                   'Train RSE': [], 'Test RSE': [], 'Train RSS': [], 'Test RSS': [],
                   'Train R^2': [], 'Test R^2': []} for model in param_grids.keys()}

# Dictionary to store best hyperparameters
best_params = {}

# Number of iterations for different train-test splits
iterations = 20

# Main loop for evaluation
for iteration in range(iterations):
    print(f"Iteration {iteration + 1}/{iterations}")

    # Split the data
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42 + iteration)

    # Iterate through models
    for model_name, grid_params in param_grids.items():
        try:
            # Define pipelines for each model
            if model_name == 'Principal Component Regression (PCR)':
                model = make_pipeline(StandardScaler(), PCA(), LinearRegression())
            elif model_name == 'Regression Splines':
                model = make_pipeline(StandardScaler(), SplineTransformer(), LinearRegression())
            elif model_name == 'Ridge Regression':
                model = make_pipeline(StandardScaler(), Ridge(max_iter=5000))
            elif model_name == 'Lasso Regression':
                model = make_pipeline(StandardScaler(), Lasso(max_iter=5000))
            elif model_name == 'Regression Tree':
                model = make_pipeline(StandardScaler(), DecisionTreeRegressor())

            # Grid search for hyperparameter tuning
            grid_search = GridSearchCV(
                model, grid_params, cv=5, scoring='neg_mean_squared_error', n_jobs=-1
            )
            grid_search.fit(X_train, y_train)

            # Get the best model and parameters
            best_model = grid_search.best_estimator_
            best_params[model_name] = grid_search.best_params_

            # Make predictions
            y_train_pred = best_model.predict(X_train)
            y_test_pred = best_model.predict(X_test)

            # Calculate metrics
            train_mse = mean_squared_error(y_train, y_train_pred)
            test_mse = mean_squared_error(y_test, y_test_pred)
            train_rmse = np.sqrt(train_mse)
            test_rmse = np.sqrt(test_mse)
            train_rss = np.sum((y_train - y_train_pred) ** 2)
            test_rss = np.sum((y_test - y_test_pred) ** 2)
            train_rse = np.sqrt(train_rss / (len(y_train) - X_train.shape[1] - 1)) if len(y_train) > X_train.shape[1] + 1 else np.nan
            test_rse = np.sqrt(test_rss / (len(y_test) - X_test.shape[1] - 1)) if len(y_test) > X_test.shape[1] + 1 else np.nan
            train_r2 = r2_score(y_train, y_train_pred)
            test_r2 = r2_score(y_test, y_test_pred)

            # Store results
            results[model_name]['Train MSE'].append(train_mse)
            results[model_name]['Test MSE'].append(test_mse)
            results[model_name]['Train RMSE'].append(train_rmse)
            results[model_name]['Test RMSE'].append(test_rmse)
            results[model_name]['Train RSE'].append(train_rse)
            results[model_name]['Test RSE'].append(test_rse)
            results[model_name]['Train RSS'].append(train_rss)
            results[model_name]['Test RSS'].append(test_rss)
            results[model_name]['Train R^2'].append(train_r2)
            results[model_name]['Test R^2'].append(test_r2)

        except Exception as e:
            print(f"Error with {model_name}: {e}")

# Calculate average results
average_results = {
    model: {metric: np.mean(scores) for metric, scores in metrics.items()}
    for model, metrics in results.items()
}

# Save and display best hyperparameters
print("Best Hyperparameters for Each Model:")
for model_name, params in best_params.items():
    print(f"{model_name}: {params}")
with open("best_params.json", "w") as f:
    json.dump(best_params, f)

# Save results to CSV
results_df = pd.DataFrame(average_results).T
results_df.to_csv("model_metrics.csv", index=True)

# Function to plot learning curves
def plot_learning_curve(model, X, y, title, cv=5):
    train_sizes, train_scores, test_scores = learning_curve(
        model, X, y, cv=cv, scoring='neg_mean_squared_error', train_sizes=np.linspace(0.1, 1.0, 10), n_jobs=-1
    )

    train_scores = -train_scores
    test_scores = -test_scores

    plt.figure(figsize=(8, 6))
    plt.plot(train_sizes, train_scores.mean(axis=1), label="Training Error", color="blue")
    plt.fill_between(train_sizes, train_scores.mean(axis=1) - train_scores.std(axis=1),
                     train_scores.mean(axis=1) + train_scores.std(axis=1), color="blue", alpha=0.2)
    plt.plot(train_sizes, test_scores.mean(axis=1), label="Validation Error", color="orange")
    plt.fill_between(train_sizes, test_scores.mean(axis=1) - test_scores.std(axis=1),
                     test_scores.mean(axis=1) + test_scores.std(axis=1), color="orange", alpha=0.2)
    plt.title(title)
    plt.xlabel("Training Set Size")
    plt.ylabel("MSE")
    plt.legend(loc="best")
    plt.grid()
    plt.show()
```

```python
from sklearn.svm import SVR

# Define SVR model
svr_model = SVR()

# Run experiments for SVR
def run_svr_experiment(X_normalized, y, n_runs=2):
    results = {'SVR': {'MSE': [], 'RMSE': [], 'RSE': [], 'RSS': [], 'R^2': []}}
    for _ in range(n_runs):
        X_train, X_test, y_train, y_test = train_test_split(X_normalized, y, test_size=0.3, random_state=None)
        svr_model.fit(X_train, y_train)
        y_pred = svr_model.predict(X_test)
        mse = mean_squared_error(y_test, y_pred)
        rmse = np.sqrt(mse)
        rss = np.sum((y_test - y_pred) ** 2)
        rse = np.sqrt(rss / (len(y_test) - X_test.shape[1] - 1))
        r2 = r2_score(y_test, y_pred)
        results['SVR']['MSE'].append(mse)
        results['SVR']['RMSE'].append(rmse)
        results['SVR']['RSE'].append(rse)
        results['SVR']['RSS'].append(rss)
        results['SVR']['R^2'].append(r2)
    return results

# Running the experiment for SVR
svr_results = run_svr_experiment(X_normalized, y)
average_svr_results = {metric: np.mean(scores) for metric, scores in svr_results['SVR'].items()}
print("SVR Results:", average_svr_results)

results_df = pd.DataFrame(average_svr_results).T
results_df.index = ['SVR']
print(results_df)
```

```python
import matplotlib.pyplot as plt
from sklearn.metrics import mean_squared_error

# Function to calculate the learning curve
def learning_curve_svr(X_normalized, y, model, train_sizes, test_size=0.3):
    train_errors = []
    test_errors = []

    for train_size in train_sizes:
        # Randomly split the dataset with the specified train size
        X_train, X_test, y_train, y_test = train_test_split(
            X_normalized, y, train_size=train_size, test_size=test_size, random_state=42
        )

        # Train the model
        model.fit(X_train, y_train)

        # Predictions on train and test sets
        y_train_pred = model.predict(X_train)
        y_test_pred = model.predict(X_test)

        # Calculate MSE for train and test sets
        train_mse = mean_squared_error(y_train, y_train_pred)
        test_mse = mean_squared_error(y_test, y_test_pred)

        train_errors.append(train_mse)
        test_errors.append(test_mse)

    return train_errors, test_errors

# Define train sizes (fraction of total data)
train_sizes = np.linspace(0.1, 0.9, 10)

# Generate learning curve data
train_errors, test_errors = learning_curve_svr(X_normalized, y, svr_model, train_sizes)

# Plot the learning curve
plt.figure(figsize=(10, 6))
plt.plot(train_sizes, train_errors, label='Training Error', marker='o')
plt.plot(train_sizes, test_errors, label='Testing Error', marker='o')
plt.title('Learning Curve for SVR')
plt.xlabel('Training Set Size (Fraction)')
plt.ylabel('Mean Squared Error')
plt.legend()
plt.grid()
plt.show()
```
